```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
import numpy as np
```

```python
fake_data = pd.read_csv('C:/Users/LENOVO/Documents/Fake_News_Detection/fake_news.csv')
fake_data['label'] = 1
```

```python
real_data = pd.read_csv('C:/Users/LENOVO/Documents/Fake_News_Detection/real_news.csv')
real_data['label'] = 0
```

```python
additional_fake_data = pd.read_csv('C:/Users/LENOVO/Documents/Fake_News_Detection/politifact_fake.csv')
additional_fake_data['label'] = 1
```

```python
additional_real_data = pd.read_csv('C:/Users/LENOVO/Documents/Fake_News_Detection/politifact_real.csv')
additional_real_data['label'] = 0
```

```python
combined_data = pd.concat([fake_data, real_data, additional_fake_data, additional_real_data])
X_train, X_test, y_train, y_test = train_test_split(
    combined_data['title'], combined_data['label'], test_size=0.2, random_state=42
)
```

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python

scaler = StandardScaler(with_mean=False)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)
```

```python

logistic_pred = logistic_model.predict(X_test)
logistic_accuracy = accuracy_score(y_test, logistic_pred)
print("Logistic Regression Accuracy:", logistic_accuracy)
```

```python

logistic_cm = confusion_matrix(y_test, logistic_pred)
labels = ['Real', 'Fake']
plt.figure(figsize=(6, 4))
sns.heatmap(logistic_cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

```python

train_sizes, train_scores, test_scores = learning_curve(logistic_model, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)
```

```python

plt.figure(figsize=(8, 6))
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.plot(train_sizes, train_mean, label='Training Accuracy', color='b')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='b', alpha=0.1)
plt.plot(train_sizes, test_mean, label='Validation Accuracy', color='g')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='g', alpha=0.1)
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve - LogisticRegression')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python

svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python

svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_accuracy)
```

```python

svm_cm = confusion_matrix(y_test, svm_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(svm_cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix')
plt.show()
```

```python

train_sizes, train_scores, test_scores = learning_curve(svm_model, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)
```

```python

plt.figure(figsize=(8, 6))
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.plot(train_sizes, train_mean, label='Training Accuracy', color='b')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='b', alpha=0.1)
plt.plot(train_sizes, test_mean, label='Validation Accuracy', color='g')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='g', alpha=0.1)
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve - SVM')
plt.legend(loc='best')
plt.grid()
plt.show()
```
